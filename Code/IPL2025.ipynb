## Team wise Win Percentage

```python
# 2. Matches won by each team
matches_won = matches['match_winner'].value_counts()
```

```python
# 3. Creating dataframe for team stats
team_stats = pd.DataFrame({
    'Matches Played': matches_played,
    'Matches Won': matches_won
}).fillna(0)

# Ensure correct datatype
team_stats['Matches Won'] = team_stats['Matches Won'].astype(int)
```

```python
# 4. Calculate the win percentage
team_stats['Win %'] = (team_stats['Matches Won']/team_stats['Matches Played']) * 100
team_stats['Win %'] = team_stats['Win %'].round(2)
```

```python
# 5. Sort BY Win %
team_stats_sorted = team_stats.sort_values(by='Win %', ascending = False)
```

```python
# Plotting
plt.figure(figsize=(10,6))
sns.set(style="whitegrid")
sns.barplot(x=team_stats_sorted['Win %'], y=team_stats_sorted.index, palette = 'coolwarm')
plt.title("Team-wise Win Percentage in IPL 2025")
plt.xlabel("Win Percentage")
plt.ylabel("Team")
plt.tight_layout()
plt.show()
```

#  How teams performed while chasing vs defending?

```python
# 1. Clean Up
matches['toss_decision'] = matches['toss_decision'].str.lower()
matches['match_winner'] = matches['match_winner'].fillna('No Result')
```

```python
# 2. Identifying which teams batted first and second
def batting_first(row):
    if row['toss_decision'] == 'bat':
        return row['toss_winner']
    else:
        return row['team1'] if row['toss_winner'] == row['team2'] else row['team2']

matches['bat_first']= matches.apply(batting_first, axis=1)
matches['bat_second']= matches.apply(lambda row:row['team1'] if row['bat_first'] == row['team2'] else row['team2'], axis = 1)
```

```python
# 3. Count wins while defending and chasing
defending_wins = matches[matches['match_winner'] == matches['bat_first']]['match_winner'].value_counts()
chasing_wins = matches[matches['match_winner'] == matches['bat_second']]['match_winner'].value_counts()
```

```python
# 4. Combine result
batting_result = pd.DataFrame({
    'Wins While Defending' : defending_wins,
    'Wins While Chasing' : chasing_wins
}).fillna(0).astype(int)
```

```python
# 5. Total Wins and Preferred strategy
batting_result['Total_Wins'] = (
    batting_result['Wins While Defending'] + batting_result['Wins While Chasing']
)
batting_result['Preferred Strategy'] = batting_result.apply(
    lambda row: 'Defending' if row['Wins While Defending'] > row['Wins While Chasing'] else 'Chasing', axis=1
)
```

```python
# 6. Show Final Results sorted by total wins
batting_result = batting_result.sort_values(by='Total_Wins', ascending=False).reset_index()
batting_result = batting_result.rename(columns={'index': 'Team'})
```

```python
# Step 7: Print the table
print("\nChasing vs Defending Analysis (IPL 2025): \n")
print(batting_result)
```

```python
# 8. Plotting
plt.figure(figsize=(12,6))
sns.set(style='whitegrid')
plot_df = batting_result.sort_values(by='Total_Wins',ascending=False) # Sorting by total Wins

# plot bar chart
bar_width = 0.4
teams = plot_df.index.tolist() 
x = range(len(teams))
plt.bar(x, plot_df['Wins While Defending'], width= bar_width, label='Defending Wins', color = 'skyblue')
plt.bar([i+ bar_width for i in x], plot_df['Wins While Chasing'],width=bar_width,label='Chasing Wins', color='salmon')
plt.xlabel("Team")
plt.ylabel("Number of Wins")
plt.title("Wins While Chasing vs Defending (IPL 2025)")
custom_team_names = [
    "RCB", "PBKS", "GT", "MI", "DC",
    "LSG", "SRH", "KKR", "RR", "CSK"
]

plt.xticks([i + bar_width / 2 for i in x], custom_team_names, rotation=45)

plt.legend()
plt.tight_layout()
plt.show()
```

## Top 10 Run Scorers in the Tournament

```python
# Creating a separate a variable for storing top scorers
batsman_stats = merged_df.groupby(['match_no','striker','batting_team']).agg(
    runs_scored = ('runs_of_bat','sum'),
    balls_faced = ('over','size') #Count of balls faced by the batsman in the match
).reset_index()
batsman_stats['strike_rate'] = (batsman_stats['runs_scored'] / batsman_stats['balls_faced']) * 100

top_batsman = batsman_stats.groupby('striker').agg(
    total_runs_scored=('runs_scored', 'sum'),
    average_runs_per_match=('runs_scored', 'mean'),
    strike_rate=('strike_rate', 'mean')
).sort_values(by='total_runs_scored', ascending=False).head(10)

#Plot the chart
plt.figure(figsize=(12,7))
top_batsman['total_runs_scored'].plot(kind = 'bar', color = 'teal')
plt.xlabel('Batsman')
plt.ylabel('Total Runs Scored')
plt.title('Top 10 Run Scorers in the Tournament')
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

```python
bowler_stats = merged_df.groupby(['match_no','bowler','bowling_team']).agg(
    runs_conceded = ('total_runs', 'sum'),
    balls_bowled = ('over','size') #Count of balls bowled by the bowler in a match
).reset_index()
bowler_stats['overs_bowled'] = bowler_stats['balls_bowled'] / 6
bowler_stats['economy_rate'] = bowler_stats['runs_conceded'] / bowler_stats['overs_bowled']

top_bowlers = bowler_stats.groupby('bowler').agg(
    total_runs_conceded = ('runs_conceded', 'sum'),
    economy_rate = ('economy_rate', 'mean'),
    balls_bowled = ('balls_bowled', 'sum')
).sort_values(by='total_runs_conceded',ascending=True).head(10)

plt.figure(figsize=(10,7))
# Sort bowlers by economy rate in ascending order for better visualisation 
top_bowlers_eco = bowler_stats.groupby('bowler').agg(
    total_runs_conceded = ('runs_conceded','sum'),
    economy_rate = ('economy_rate', 'mean'),
    balls_bowled = ('balls_bowled','sum')
).sort_values(by = 'economy_rate', ascending = True).head(10)

top_bowlers_eco['economy_rate'].plot(kind = 'bar', color = 'orange')
plt.xlabel('Bowler')
plt.ylabel('Economy Rate')
plt.title('Top 10 Bowler By Economy Rate')
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()
plt.show()
```

```python
# Valid dismissal types that count as bowler's wicket
valid_dismissals = ['nan', 'caught', 'bowled', 'stumped', 'lbw', 'runout', 'retired out',
       'retired hurt', 'hit wicket']

# Filter only valid wickets
wickets = deliveries[deliveries['wicket_type'].isin(valid_dismissals)]

# Count wickets by bowler
top_bowlers = wickets['bowler'].value_counts().reset_index()
top_bowlers.columns = ['Bowler', 'Wickets']
top_5 = top_bowlers.head(5)

# Plot the bar chart
plt.figure(figsize=(10,8))
bars = plt.bar(top_5['Bowler'], top_5['Wickets'], color = 'mediumseagreen')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.5, int(yval), ha = 'center', va = 'center')

plt.title('Top 5 Wicket Takers in IPL 2025')
plt.xlabel('Bowlers')
plt.ylabel('No. of Wickets')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
